```python
def peak_analysis(entry, rbf_type = 'Gaussian', data_used = 'Combined Re-Im Data', induct_used = 1, der_used = '1st order', cv_type = 'GCV', shape_control = 'FWHM Coefficient', coeff = 0.5, peak_num = 3):
    
    """
       This function enables to identify the DRT peaks.
       Inputs:
            entry: An EIS spectrum
            rbf_type: Discretization function
            data_used: Part of the EIS spectrum used for regularization
            induct_used: Treatment of the inductance part
            der_used: Order of the derivative considered for the M matrix
            cv_type: Regularization method used to select the regularization parameter for ridge regression 
            shape_control: Option for controlling the shape of the radial basis function (RBF) 
            coeff: Magnitude of the shape control
            peak_num: Number of peaks in the DRT spectrum         
    """
    
    simple_run(entry, rbf_type, data_used, induct_used, der_used, cv_type, shape_control, coeff) # to get the DRT
    
    # the desired number of peaks is entered by the user
    entry.N_peak = np.int_(peak_num) # number of peaks inputted
    
    if entry.N_peak < 1:
        
        entry.N_peak = 1
           
    # Step 1 compute the first inital value
    
    entry.index_peak_max = np.argmax(entry.x) # position of the highest DRT peak
    entry.peak_value_max = entry.x[entry.index_peak_max] #value of the highest DRT peak
    
    entry.tau_vec = 1.0 / entry.freq # vector of log timescales of size N_freqs
    
    entry.log_tau_mu_max = np.log(entry.tau_vec[entry.index_peak_max]) # log timescale of the highest DRT peak
    
    # avoid potential numerical issue when log_tau_mu_max is close to zeros
    eps = np.finfo(float).eps
    if np.abs(np.exp(entry.log_tau_mu_max) - 1) < eps:
        
        entry.log_tau_mu_max = np.log(entry.tau(entry.index_peak_max) + eps)
    
    # difference in collocation points
    
    entry.sigma_vec = np.mean(np.diff(np.log(1./entry.freq))) # average value of differences of log timescales (scalar)
    
    # construct the inital value:
    
    # p_init takes the form of [R_1, log_tau_mu_1, sigma_1]
    
    entry.p_ref = [entry.peak_value_max, entry.log_tau_mu_max, entry.sigma_vec]
    
    entry.p_init = np.ones_like(entry.p_ref)
    
    entry.lb = [0, np.min(np.log(entry.tau_fine)), 0] 
    entry.ub = [np.inf, np.max(np.log(entry.tau_fine)), np.inf]
    
    # Step 2 for loop through N
    
    # minimize the sum of squared residual betwen gamma and gamma_peak
    
    for n in range(1, entry.N_peak +1):
        
        print('working on peak', n)
        
        sq_residual_fct = lambda p: np.linalg.norm(gf.gauss_fct(entry.tau_vec, entry.p_ref, p) - (entry.x).T)**2 
        
        options = {'algorithm': 'active-set', 'display': True, 'tol': 1e-15, 'maxiter': 2000, 'maxfun': int(1e5)}
        
        result = minimize(sq_residual_fct, entry.p_init, method='SLSQP', bounds=Bounds(entry.lb, entry.ub), options=options)
        
        entry.p_fit = result.x

        # compute the DRT difference between the ridge DRT and peak DRT
        
        entry.abs_residual = np.abs(gf.gauss_fct(entry.tau_vec, entry.p_ref, entry.p_fit) - (entry.x).T)
        
        # guess new peak position
        
        entry.index_peak_max_temp = np.argmax(entry.abs_residual)
        
        entry.peak_value_temp = entry.x[entry.index_peak_max_temp]
        entry.log_tau_mu_temp =  np.log(entry.tau_vec[entry.index_peak_max_temp])
        
        if np.abs(np.exp(entry.log_tau_mu_temp) - 1) < eps: 
            
            entry.log_tau_mu_temp = np.log(entry.tau_vec(entry.index_peak_max_temp) + eps)
   
        entry.sigma_temp = np.mean(np.diff(np.log(1.0 / entry.freq))) # I think this is sigma_0_vec (line 37)
    
        entry.p_ref_temp = np.array([entry.peak_value_temp, entry.log_tau_mu_temp, entry.sigma_temp])
        
        entry.p_init_temp = np.ones(np.shape(entry.p_ref_temp))
        
        entry.p_fit_temp = np.ones(np.shape(entry.p_ref_temp))
        
        entry.lb_temp = np.array([0, np.argmin(np.log(entry.tau_vec)), 0])
        
        entry.ub_temp = np.array([np.inf,np.argmax(np.log(entry.tau_vec)), np.inf])
        
        if n != entry.N_peak: # where is the else counterpart
            # p_ref = np.array([np.multiply(p_ref,p_fit),np.multiply(p_ref_temp,p_fit_temp)])
            
            entry.p_ref = [np.dot(entry.p_ref, entry.p_fit), np.dot(entry.p_ref_temp, entry.p_fit_temp)]
            
            entry.p_init = np.ones(np.shape(entry.p_ref))
            
            entry.lb = np.array([entry.lb, entry.lb_temp])
            
            entry.ub = np.array([entry.ub, entry.ub_temp])
        
    entry.p_result = np.reshape(np.multiply(entry.p_ref, entry.p_fit), tuple(np.array([3, entry.N_peak])), order="F")
    
    entry.gamma_gauss = gf.gauss_fct(entry.tau_vec, entry.p_ref, entry.p_fit)
    
    # Step 4 for loop convert gaussian function to gamma_mat
    
    entry.gamma_gauss_mat = np.zeros((np.asarray(entry.tau_vec).size, entry.N_peak))
    
    entry.g_gauss_mat = np.zeros(((entry.tau_vec).size, entry.N_peak))
    
    for i in np.arange(1, entry.N_peak+1).reshape(-1):
        
        R_0 = entry.p_result(1,i)
        
        mu_log_tau = entry.p_result(2,i)
        
        sigma = entry.p_result(3,i)
        
        entry.gamma_gauss_mat[:,i] = R_0 * np.exp(- (np.log(1.0 / entry.freq) - mu_log_tau) ** 2.0 / (2 * sigma) ** 2)
        
        entry.g_gauss_mat[:,i] = np.multiply(entry.gamma_gauss_mat[:,i],entry.freq)
    
    # change the method that we have conducted peak analysis.
    entry.method = 'peak'
    
    return entry
```
